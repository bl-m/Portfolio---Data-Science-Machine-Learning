# Business Problem 



The data set, Online Retail II, includes sales of an online store based in the UK from 01/12/2008 to 09/12/2011.

The company sells promotional products, products like souvenirs, etc.

The business problem is that one e-commercial firm try to determine its strategy based on the customer segment by grouping them into different segments.
RFM analysis technique will be used to to determine these custumer segments.
RFM considers recency, frequency and monetary values for each customer. Grouping them into different customer segments for campaign targeting is the premise.
Online Reatail II dataset from UK-based firm, covering purchases from 01/12/2009 to 09/12/2010, will be used for the scope of this study. Only data icluding 2010 and 2011 will be used.

Variables:

InvoiceNo: Unique number for each transaction
StockCode: Unique number for each item
Description: Product name
Quantity: Number of items sold
InvoiceDate: Date of Invoice
UnitPrice: Price for each item
CostomerID: Unique costomer ID
Country: Country where costomers reside


https://archive.ics.uci.edu/ml/datasets/Online+Retail+II

# Frequency

```python
temp_df = df.groupby(["Customer ID","Invoice"]).agg({"Invoice":"count"})
```

```python
temp_df.head()
```

```python
temp_df.groupby("Customer ID").agg({"Invoice":"sum"}).head()
```

```python
freq_df = temp_df.groupby("Customer ID").agg({"Invoice":"sum"})
freq_df.rename(columns={"Invoice": "Frequency"}, inplace = True)
freq_df.head()
```

# Monetary

```python
monetary_df = df.groupby("Customer ID").agg({"TotalPrice":"sum"})
```

```python
monetary_df.head()
```

```python
monetary_df.rename(columns={"TotalPrice": "Monetary"}, inplace = True)
```

```python
print(recency_df.shape,freq_df.shape,monetary_df.shape)
```

```python
rfm = pd.concat([recency_df, freq_df, monetary_df],  axis=1)
```

```python
rfm.head()
```

```python
rfm["RecencyScore"] = pd.qcut(rfm['Recency'], 5, labels = [5, 4, 3, 2, 1])
```

```python
rfm["FrequencyScore"] = pd.qcut(rfm['Frequency'], 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm["MonetaryScore"] = pd.qcut(rfm['Monetary'], 5, labels = [1, 2, 3, 4, 5])
```

```python
rfm.head()
```

```python
(rfm['RecencyScore'].astype(str) + 
 rfm['FrequencyScore'].astype(str) + 
 rfm['MonetaryScore'].astype(str)).head()
```

```python
rfm["RFM_SCORE"] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str) + rfm['MonetaryScore'].astype(str)
```

```python
rfm.head()
```

```python
rfm.describe().T
```

```python
rfm[rfm["RFM_SCORE"] == "555"].head()
```

```python
rfm[rfm["RFM_SCORE"] == "111"].head()
```

```python
seg_map = {
    r'[1-2][1-2]': 'Hibernating',
    r'[1-2][3-4]': 'At Risk',
    r'[1-2]5': 'Can\'t Loose',
    r'3[1-2]': 'About to Sleep',
    r'33': 'Need Attention',
    r'[3-4][4-5]': 'Loyal Customers',
    r'41': 'Promising',
    r'51': 'New Customers',
    r'[4-5][2-3]': 'Potential Loyalists',
    r'5[4-5]': 'Champions'
}
```

```python
rfm['Segment'] = rfm['RecencyScore'].astype(str) + rfm['FrequencyScore'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(seg_map, regex=True)
rfm.head()
```

```python
rfm[["Segment", "Recency","Frequency","Monetary"]].groupby("Segment").agg(["mean","count"])
```
